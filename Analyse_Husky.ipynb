```python
def classify_row(df,node,probability=True):
    #sevre para chegarmos as folhas
    #s serve para uma linha/amostra de dados serve 
    if isinstance(node,m2.Leaf):
        if probability:
            return node.probability
        else:
            return node.predictions
    if node.question.match(df):
        return classify_row(df,node.true_branch,probability)
    else:
        return classify_row(df,node.false_branch,probability)
    
def test_tree(df,tree,var=1,tree_limite=60,proba=False):
    #testa uma rvore, sabendo que a var s  considerada 1 se probabilidade>=limite
    #devolve uma lista com os valores q previu
    #proba diz se queremos a probability, ou usar o limite
    size=len(df.index)
    predicted=[] #valores previstos vo estar em lista
    for i in range(size):
        row=pd.DataFrame(df.iloc[i,:]).T
        d_proba=classify_row(row,tree)
        if not(proba):
            if d_proba[var]>=tree_limite:
                predicted+=[var]
            else:
                predicted+=[1-var] #se var=0,1 e se var=1,0
        else:
            predicted+=[d_proba[var]]
    #print('red',predicted)
    return predicted

def forest_classify(df,forest,tree_limite=60):
    #devolve lista de dicionrios em que cada dict refer-se a uma amostra/linha
    #e os valores de cada dict referem-se as previses de cada rvore em 
    #relao a essa amostra
    rows=len(df.index)
    bigout=[]
    bigout2=[]
    for i in range(rows):
        print(i,'i')
        output={0:0,1:0}
        output2=0
        row=pd.DataFrame(df.iloc[i,:]).T
        trees=forest.size()
        for n in range(trees):
            tree=forest.get_tree(n)
            
            val=test_tree(row,tree,tree_limite=tree_limite)[0] #previses 0/1
            
            val2=test_tree(row,tree,tree_limite=tree_limite,proba=True)[0] #score
            output2+=val2 #score
            output[val]+=1 #previses 0/1
            
        bigout+=[output]
        
        bigout2+=[output2]
    return bigout,bigout2

def forest_predictions(df, forest, var=1, forest_limite=60,tree_limite=60):
    #devolve uma lista com as previses da floresta
    #impondo um limite (percentagem) a uma varivel
    #(s dizemos que se trata dessa varivel se a percentagem de probabilidade)
    #for igual ou superior
    final=[]
    classi=forest_classify(df,forest,tree_limite=tree_limite)
    #print(classi)
    for i in classi:
        print('len',len(final))
        prob=round((i[var]/(forest.size()))*100,2)
        if prob>=forest_limite:
            final+=[var]
        else:
            final+=[1-var]
    return final

def forest_sum_trees(df, forest, var=1, tree_limite=60):
    #devolve a soma das probas de cada rvores (O Score das rvores)
    'a'
```

```python
def classify_row(df,node,probability=True):
    #sevre para chegarmos as folhas
    #s serve para uma linha/amostra de dados serve 
    if isinstance(node,Leaf):
        if probability:
            return node.probability
        else:
            return node.predictions
    if node.question.match(df):
        return classify_row(df,node.true_branch,probability)
    else:
        return classify_row(df,node.false_branch,probability)
    
def test_tree(df,tree,var=1,tree_limite=60,proba=False):
    #testa uma rvore, sabendo que a var s  considerada 1 se probabilidade>=limite
    #devolve uma lista com os valores q previu
    #proba diz se queremos a probability, ou usar o limite
    size=len(df.index)
    predicted=[] #valores previstos vo estar em lista
    for i in range(size):
        row=pd.DataFrame(df.iloc[i,:]).T
        d_proba=classify_row(row,tree)
        if not(proba):
            if d_proba[var]>=tree_limite:
                predicted+=[var]
            else:
                predicted+=[1-var] #se var=0,1 e se var=1,0
        else:
            predicted+=[d_proba[var]]
    #print('red',predicted)
    return predicted
```
