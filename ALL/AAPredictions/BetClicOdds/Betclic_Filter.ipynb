```python
import pandas as pd
import numpy as np
import os
import datetime as dt
```

```python
def remove_blank(string_list):
    s_list=[]
    for s in string_list:
        if s!='':
            s_list.append(s)
    return s_list


leagues_dix={
    'Itália - Serie A':'Serie_A',
    'Espanha - La Liga': 'La_Liga',
    'Inglaterra - Premier Lg.': 'Premier_League', 
    'Portugal - Primeira Liga':'Primeira_Liga' #não vou usar por agr, mas vou guardando as odds
}

def filtery(df,leagues_dix=leagues_dix):
    df = df.drop(['web-scraper-order','web-scraper-start-url','games','games-href'],axis=1)
    leag=[]
    date=[]
    time=[]
    ht=[]
    at=[]
    remove_odds=[]
    for l in range(len(df)):
        teams=df.iloc[l]['Teams']
        lig=df.iloc[l]['League']
        dte=df.iloc[l]['Date']
        if lig not in leagues_dix:
            remove_odds+=[l]
            print(lig, 'não está no league_dix')
            continue
        else:
            leag+=[leagues_dix[lig]]
            htim,atim = teams.split(' - ')
            ht+=[htim]
            at+=[atim]
            dia,timez = remove_blank(dte.replace('\n',' ').split(' ')) 
            time+=[timez]
            if dia=='Hoje':
                date+=[str(dt.date.today())]
            elif dia=='Amanhã':
                date+=[str(dt.date.today()+dt.timedelta(1))]
            else:
                if '/' not in dia:
                    print('Oops, problemo avec les datas')
                    raise('oo')
                date+=[dt.datetime.strptime(dia,'%d/%m/%Y')]
    for c in df.columns:
        if 'ODD' in c:
            vars()[c] = np.delete(np.array(list(df.iloc[:][c])),remove_odds)
    df_new = pd.DataFrame({})
    df_new['League']=leag
    df_new['Date (YYYY-MM-DD)']=date
    df_new['Time']=time
    df_new['HT']=ht
    df_new['AT']=at
    for c in df.columns:
        if 'ODD' in c:
            df_new[c]=vars()[c]
    return df_new.sort_values(['Date (YYYY-MM-DD)','Time'],ascending=[True,True]).reset_index(drop=True)
```

```python
df1 = pd.read_csv('betclic_top_europa5.csv')
```

```python
df_new=filtery(df1)
```

```python
df_new.to_excel('Repos_BetClic5.xlsx',index=False)
```
